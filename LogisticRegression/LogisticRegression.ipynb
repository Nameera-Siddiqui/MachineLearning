```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```
df  = pd.read_csv("insurance_data.csv")
df.head(5)
```

```
plt.scatter(df['age'], df['bought_insurance'], marker='.',color='red')
```

```
x_train,x_test,y_train,y_test = train_test_split(df[['age']] , df.bought_insurance , test_size=0.2)
x_test
```

```
model=LogisticRegression()
model.fit(x_train,y_train)
```

```
model.predict(x_test)
```

```
y_test
```

```
model.score(x_test,y_test)
```

```
model.predict_proba(x_test) #This tells how the model is making decision (Not Buy vs Buy) > value wins
```

```
model.predict(np.array([56]).reshape(-1,1))
```

```
model=LogisticRegression()
model.fit(df[['age']],df['bought_insurance'])
#Generating age for plotting the curve
age_range = np.linspace(df.age.min(),df.age.max(),300).reshape(-1,1)
predicted_probs = model.predict_proba(age_range)[:,1]
plt.figure(figsize=(10,7))
plt.scatter(df[['age']],df.bought_insurance,marker='.',color='green', label = 'Data Points', alpha = 0.6)
plt.plot(age_range,predicted_probs, color='red', label = 'Logistic Regression Curve', linewidth = 2)
plt.xlabel('Age')
plt.ylabel('Probability of Buying Insurance')
plt.title('Logistic Regression Curve for Age vs Insurane Purchase')
plt.legend()
plt.grid(True)
plt.show()
```
