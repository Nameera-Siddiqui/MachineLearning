```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df =  pd.read_csv("Salary_dataset.csv")
df
```

```python
%matplotlib inline
plt.xlabel('Experience(Years)')
plt.ylabel('Salary (US $)')
plt.scatter(df.YearsExperience,df.Salary, color='red', marker= '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['YearsExperience']], df.Salary)
```

```python
# onverting into 2D array
pred_price = reg.predict([[7.5]])
print("The predicted price for  input Year of exp is = $", pred_price)
```

```python
pred_price = reg.predict([[12]])
print("The predicted price for  input Year of exp is = $", pred_price)
```

```python
%matplotlib inline
plt.xlabel('Experience (Y) ', fontsize=15)
plt.ylabel('Salary ($)' , fontsize=15)
plt.scatter(df.YearsExperience, df.Salary , color='green', marker =  '+')

plt.plot(df.YearsExperience, reg.predict(df[['YearsExperience']]), color='purple')
```
