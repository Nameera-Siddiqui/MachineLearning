```
#One Hot Endoing and Creating Dummy Variable
import numpy as np
import pandas as pd
```

```
df = pd.read_csv("homeprices2.csv")
df
```

```
# Pandas has a dummy method to create dummy variables for our town
dummies = pd.get_dummies(df.town, dtype = int)
dummies
```

```
# Pandas concat method concatenates two-column, write first table name, then second table name and their axis of mergin
merges = pd.concat([df,dummies] , axis='columns')
merges
```

```
# We will drop the town(Because we have corresponding binary for it) and one of the dummy variable columns.
#The reason to drop them is that town column will not help us with linear prediction
#Same as that when we are dealing with the dummy variables we have to drop one column
# to save ourselves form dummy variable trap.

final = merges.drop(['town','robinsville'], axis ='columns')
final
```

```
#Applying Linear Regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```
# we will drop price because this is the output we want to predict by the model
x = final.drop(['price'] , axis = 'columns')
x
```

```
y = final.price
y
```

```
# when we use fit it means we are going to train our model
model.fit(x,y)
```

```
#Now how this works is that 10 represent monroe township, 01 represents west windsor 	and  00 represent robinsville
predicted_PP = model.predict([[2700,1,0]]) # area, 1/0 , 0/1
print("The predicted price of house for Monroe township is = ", predicted_PP)
```

```
predicted_PP = model.predict([[2700,0,0]])
print("The predicted price of house for robinsville is = ", predicted_PP)
```

```
predicted_PP = model.predict([[2700,0,1]])
print("The predicted price of house for west windsor is = ", predicted_PP)
```
